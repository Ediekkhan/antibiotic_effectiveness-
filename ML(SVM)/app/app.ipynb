```python
from flask import Flask, render_template, request
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import joblib

app = Flask(__name__)

# Load the trained model, scaler, and encoding columns
model, scaler, diagnosis_columns, drugs_columns = joblib.load('/home/Kanex/Documents/dr An/ML(SVM)/Test_train.ipynb')

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    if request.method == 'POST':
        diagnosis = request.form['diagnosis']
        drugs = request.form['drugs']

        # Preprocess the input features
        features = pd.DataFrame({
            'diagnosis': [diagnosis],
            'drugs': [drugs]
        })

        # One-hot encode the Diagnosis column
        diagnosis_encoded = pd.get_dummies(features['diagnosis'], columns=diagnosis_columns)

        # One-hot encode the Drugs column
        drugs_encoded = pd.get_dummies(features['drugs'], columns=drugs_columns, drop_first=True)

        # Concatenate the encoded columns
        input_features = pd.concat([diagnosis_encoded, drugs_encoded], axis=1)

        # Scale the features using the same scaler used during training
        input_features_scaled = scaler.transform(input_features)

        # Make prediction
        prediction = model.predict(input_features_scaled)

        return render_template('result.html', prediction=prediction[0])

if __name__ == '__main__':
    app.run(debug=True)
```
